```python
# Question 4
#Create a new column “Title”, which will contain the titles of the passengers such as Mr., Mrs., etc
import re
df_Train['Title'] = df_Train.Name.apply(lambda x: re.search(' ([A-Z][a-z]+)\.', x).group(1))
```

```python
# 4)- There should be five titles: Mr, Mrs, Master, Miss, and Special. 
#(Hint: Use regular expression to replace the titles)
df_Train['Title'] = df_Train['Title'].replace({'Mlle':'Miss', 'Mme':'Mrs', 'Ms':'Miss', 'Mr.':'Mr', 'Master.':'Master'})
df_Train['Title'] = df_Train['Title'].replace(['Rev', 'Sir', 'Major'],'Special')
df_Train
```

```python
# Question 5
#Impute the missing values using .fillna() method.
data.isnull().sum()
```

```python
data.Age=data.Age.fillna(data.Age.mean())
data.Fare=data.Fare.fillna(data.Fare.mean())
data.Embarked=data.Embarked.fillna('S')
data.head()
```

```python
df_Train.Age=df_Train.fillna(df_Train.Age.mean())
df_Train.Fare=df_Train.Fare.fillna(df_Train.Fare.mean())
df_Train.isnull().sum()
```

```python
# Question 6
#Create two new columns “Categorical_age” and “Categorical_fare”, by using the pandas function qcut() 
#to bin your numerical data.
df_Train['Categorical_age']=pd.qcut(df_Train.Age,q=4,labels=False)
df_Train['Categorical_fare']=pd.qcut(df_Train.Fare,q=4,labels=False)
df_Train.tail()
```

```python
# Question 8
#Transform all variables into numeric ones using get_dummies().
df_Train_dum = pd.get_dummies(df_Train, drop_first=True)
df_Train_dum.head()
```

```python
# Question 9
#Now we need training set and testing set. Use train_test_split() to split your dataset. 
# X(Features) will be the dataset values without the column “Survived” and y(Targets) will be “Survived” column values.
X = df_Train_dum.drop(['Survived'], axis = 1).values
y = df_Train_dum['Survived'].values
```

```python
#Splitting the model
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
# Question 10
#Use logistic regression by passing your X_train and y_train.
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train, y_train)
```

```python
#Prediction
test_pred =lr.predict(X_test)
```

```python
result=pd.DataFrame({'Actual':y_test,'Predicted':test_pred})
print(result)
```

```python
# Question 11
#Print the accuracy score.
from sklearn.metrics import accuracy_score
accuracy_score = lr.score(X_test, y_test)
print("The accuracy score is : {}".format(accuracy_score))
```
