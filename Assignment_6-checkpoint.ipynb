```python
# Question 1
#Visualisation of Weather Dataset

#importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
plt.style.use('classic')
%matplotlib inline
```

```python
#Loadind Dataset
df = pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv")
df.head()
```

```python
#Univariate Data Plotting
sns.distplot(df['MinTemp'], hist=True, kde=True, rug=False, color='Teal')
plt.show()
```

```python
plt.close()
```

```python
plt.subplot(221)
sns.distplot(df['MaxTemp'], color='red')
plt.subplot(222)
sns.distplot(df['MinTemp'], color='green')
plt.show()
```

```python
plt.close()
```

```python
#Kernel Density Estimation - KDE
sns.kdeplot(df['Humidity3pm'], shade=True, color='yellow')
plt.show()
```

```python
plt.close()
```

```python
plt.subplot(211)
sns.kdeplot(df['Pressure9am'], bw=100, label='bin, size=100')
plt.legend();
plt.show()
```

```python
plt.close()
```

```python
#Bivariate Data Plotting
#plotting Scatterplot
sns.jointplot(x='Temp9am', y='Temp3pm', data=df, color='Fuchsia')
plt.show()
```

```python
plt.close()
```

```python
#KDE for Bivariate Data
sns.jointplot(x=df['Evaporation'], y=df['Sunshine'], kind='hex', color='lime')
plt.show()
```

```python
plt.close()
```

```python
#Categorical - Stripplot
sns.stripplot(x='MinTemp', y='MaxTemp', data=df)
plt.show()
```

```python
plt.close()
```

```python
#Distribution of observation within categories
sns.swarmplot(x='Rainfall', y='Sunshine', data=df)
plt.show()
```
